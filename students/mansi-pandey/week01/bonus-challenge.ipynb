## 1Ô∏è‚É£: Create dataset

```python

import numpy as np
import matplotlib.pyplot as plt

# dataset: Hours of practice vs Skill level
practice_hours = np.array([1, 2, 3, 4, 5, 6])
skill_level = np.array([2, 4, 5, 7, 9, 11])
```

---

## 2Ô∏è‚É£: Plot the points

```python
plt.figure(figsize=(8, 5))
plt.plot(practice_hours, skill_level, linestyle='--', color='blue')
plt.scatter(practice_hours, skill_level, color='green', s=100, label='Actual Data')
plt.xlabel('Hours of Practice')
plt.ylabel('Skill Level')
plt.title('Practice vs Skill Level')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

___

## 3Ô∏è‚É£: Find the best line (slope m and intercept b)

```python
# calculate means
mean_x = practice_hours.mean()
mean_y = skill_level.mean()

# Best slope (m)
numerator = np.sum((practice_hours - mean_x) * (skill_level - mean_y))
denominator = np.sum((practice_hours - mean_x)**2)
m = numerator / denominator

# Best intercept (b)
b = mean_y - m * mean_x

print(f"Best slope (m): {m}")
print(f"Best intercept (b): {b}")
print(f"Best line equation: y = {m}x + {b}")
```

___

## 4Ô∏è‚É£: Calculate predicted values and plot line


```python
predicted_skill = m * practice_hours + b

plt.figure(figsize=(8, 5))
plt.scatter(practice_hours, skill_level, color='blue', s=100, label='Actual Data')
plt.plot(practice_hours, predicted_skill, color='green', linewidth=2, label=f'Best Line (y={m:.2f}x+{b:.2f})')
plt.xlabel('Hours of Practice')
plt.ylabel('Skill Level')
plt.title('Best Fitting Line')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

___

## 5Ô∏è‚É£: Make a prediction

```python
# Predict skill level for 7 hours of practice

hours_to_predict = 7

predicted_value = m * hours_to_predict + b

print(f"Predicted skill level for {hours_to_predict} hours of practice: {predicted_value:.2f}")
```
